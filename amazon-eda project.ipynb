# DATA CLEANING

```python
%pip install pandas 
```

```python
import pandas  as  pd 
```

```python
df=pd.read_csv("AMAZON.CSV")
df.head()
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes 
```

```python
df.info() 
```

```python
df.isnull().sum() 
```

```python
df["rating_count"] = (
    df["rating_count"]
    .astype(str)
    .str.replace(",","", regex=False)
)
```

```python
df["rating_count"]=pd.to_numeric(df["rating_count"],errors="coerce")
```

```python
df["rating_count"]=df["rating_count"].fillna(df["rating_count"].mean())
```

```python
df["rating_count"].isnull().sum()
```

```python
df["rating"]=pd.to_numeric(df["rating"],errors="coerce")
```

```python
df["rating"]=df["rating"].fillna(df["rating"].mean())
```

```python
df["rating"].isnull().sum()
```

```python
df.duplicated().sum()#it  finds number of duplicate sets in a row
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.columns
```

```python
#removing unnecessary columns
df = df.drop(columns=[
    'product_id',
    'user_id',
    'user_name',
    'review_id',
    'img_link',
    'product_link'
])
```

```python
df.columns
```

```python
#cleaning columns
df["actual_price"] = (
    df["actual_price"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",","", regex=False)
)

df["actual_price"] = pd.to_numeric(
    df["actual_price"],
    errors="coerce"
)
```

```python
df["discounted_price"] = (
    df["discounted_price"]
    .astype(str)
    .str.replace("₹","", regex=False)
    .str.replace(",", "", regex=False)
)

df["discounted_price"] = pd.to_numeric(
    df["discounted_price"],
    errors="coerce"
)
```

```python
df["rating"] = (
    df["rating"]
    .astype(str)
    .str.replace("₹","", regex=False)
    .str.replace(",","", regex=False)
)

df["rating"] = pd.to_numeric(
    df["rating"],
    errors="coerce"
)
```

```python
df["discount_percentage"] = (
    df["discount_percentage"]
    .astype(str)
    .str.replace('%', "", regex=False)
)

df["discount_percentage"] = pd.to_numeric(
    df["discount_percentage"],
    errors="coerce"
)
```

```python
corr = df[["discounted_price",
           "actual_price",
           "discount_percentage",
           "rating",
           "rating_count"]].corr() #it finds relationship between numerical values

sns.heatmap(corr,
            annot=True,
            cmap="coolwarm")
```

# Insight

### Product pricing factors are closely connected, but customer ratings are influenced more by product quality and user experience than by discounts.
